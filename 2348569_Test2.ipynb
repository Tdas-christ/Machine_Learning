# **Apply K-means clustering algorithm for clusters ranging from 2 to 10 for the dataset generated. Visualize all the clusters and centre of the clusters.**

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
scaled_df = sc.fit_transform(df)
```

```python
scaled_df
```

```python
from sklearn.cluster import KMeans

for k in range(2, 11):
  kmeans = KMeans(n_clusters=k, random_state=42)
  df['Cluster'+ str(k)] = kmeans.fit_predict(scaled_df)
```

```python
df.head()
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
df_pca = pca.fit_transform(scaled_df)
for k in range(2, 11):
  kmeans = KMeans(n_clusters=k, random_state=42)
  cluster_labels = kmeans.fit_predict(df_pca)
  plt.title(f'Clusters (K={k})')

  plt.scatter(df_pca[:, 0], df_pca[:, 1], c=cluster_labels, cmap='viridis')
  plt.xlabel('Feature 1')
  plt.ylabel('Feature 2')
  # Displaying cluster centers
  plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], c='red', marker='X', s=200, label='Centroids')
  plt.legend()
  plt.show()
```

# **Print number of samples in every cluster (ranging from 2 to 10)**

```python
for i in range(2,11):
  print(df['Cluster'+ str(i)].value_counts)
```

# **Find within cluster sum of square distance for the clusters created and plot using suitable graph**

```python
inertia_values = []
#here the inertia is nothing but the Within Cluster Sum of Squares
for k in range(2,11):
  kmeans = KMeans(n_clusters=k, random_state=42)
  kmeans.fit(scaled_df)
  inertia_values.append(kmeans.inertia_)
```

```python
plt.figure(figsize=(10, 6))
plt.plot(range(2,11), inertia_values, marker='o')
plt.title('Number of clusters against WCSS')
plt.xlabel('Number of Clusters (K)')
plt.ylabel('WCSS')
plt.show()
```

```python
#from the above plot, we can say that our optimal k value is 2
opt_k = 2
```

# **Find silhouette score of all the clusters and plot using suitable graph**

```python
from sklearn.metrics import silhouette_score
silhouette=[]
for k in range(2,11):
  kmeans = KMeans(n_clusters=k, random_state=42)
  cluster_labels = kmeans.fit_predict(df_pca)
  silhouette.append(silhouette_score(scaled_df, cluster_labels))
```

```python
k=2
for i in range(0,len(silhouette)):
  print(f"The Silhouette score for Cluster {k} is ", silhouette[i])
  k=k+1
```

```python
plt.figure(figsize=(10, 6))
plt.plot(range(2,11), silhouette, marker='o')
plt.title('Silhouette score of the clusters')
plt.xlabel('Number of Clusters (K)')
plt.ylabel('Silhouette Score')
plt.show()
```

**Write a python script to create a perceptron to
demonstrate the following
o AND operation (3 Marks)
o OR operation (2 Mark)
o XOR operation (2 Mark)
Use two input neurons and one output neuron,
Assign weights as follows and compute the value
W1=1.5, w2=0.5, w0 = -0.5 (w0 is b in the figure)
o Display the error for each function (2 Marks)**

```python
def unitStep(v):
    if v >= 0:
        return 1
    else:
        return 0

# designing the Perceptron Model
def perceptronModel(x, w, b):
    v = np.dot(w, x) + b
    y = unitStep(v)
    return y
```

```python
# AND Logic Function
def AND_GATE(x):
    w = np.array([1.5, 0.5])
    b = -0.5
    return perceptronModel(x, w, b)
```

```python
#OR Logic function
def OR_GATE(x):
    w = np.array([1.5, 0.5])
    b = -0.5
    return perceptronModel(x, w, b)
```

```python
# NOT logic function
def NOT_GATE(x):
    wNOT = -1
    bNOT = 0.5
    return perceptronModel(x, wNOT, bNOT)
```

```python
#XOR logic function
def XOR_GATE(x):
    y1 = AND_GATE(x)
    y2 = OR_GATE(x)
    y3 = NOT_GATE(y1)
    final_x = np.array([y2, y3])
    result = AND_GATE(final_x)
    return result
```

```python
# testing the Perceptron Model
t1 = np.array([0, 1])
t2 = np.array([1, 1])
t3 = np.array([0, 0])
t4 = np.array([1, 0])
```

```python
print("AND(0,1) = {}".format(AND_GATE(t1)))
print("AND(1,1) = {}".format(AND_GATE(t2)))
print("AND(0,0) = {}".format(AND_GATE(t3)))
print("AND(1,0) = {}".format(AND_GATE(t4)))
```

```python
print("OR(0,1) = {}".format(OR_GATE(t1)))
print("OR(1,1) = {}".format(OR_GATE(t2)))
print("OR(0,0) = {}".format(OR_GATE(t3)))
print("OR(1,0) = {}".format(OR_GATE(t4)))
```

```python
print("XOR(0,1) = {}".format(XOR_GATE(t1)))
print("XOR(1,1) = {}".format(XOR_GATE(t2)))
print("XOR(0,0) = {}".format(XOR_GATE(t3)))
print("XOR(1,0) = {}".format(XOR_GATE(t4)))
```

As we can see, only the OR GATE works properly witht he assigned w and b values. We need to later the values of w and b for the AND gate and XOR gate in order to get correct output.
